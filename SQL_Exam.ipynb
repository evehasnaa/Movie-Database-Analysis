```python
%%sql
-- select m.title, m.movie_id,genre_name ,kw.keyword_name
-- 55 without distinct
select count( m.movie_id) as Thriller_love_count
from movies as m
inner join genremap as g
on m.movie_id=g.movie_id
-- movie with genremap
inner join genres as gen 
on gen.genre_id= g.genre_id
-- genres with genremap
inner join keywordmap k
-- movie with keywordmap
on m.movie_id=k.movie_id
inner join keywords kw
-- keywordmap with keyword 
on k.keyword_id=kw.keyword_id
where genre_name like 'Thriller' and 
kw.keyword_name like '%love%'
order by m.movie_id
limit 20;
 
```

```python
%%sql
-- join movies with genremap and genre taples 
-- select m.title, m.movie_id,genre_name ,kw.keyword_name
-- 48 with distinct
select count(distinct m.movie_id) as Thriller_love_count
from movies as m
inner join genremap as g
on m.movie_id=g.movie_id
-- movie with genremap
inner join genres as gen 
on gen.genre_id= g.genre_id
-- genres with genremap
inner join keywordmap k
-- movie with keywordmap
on m.movie_id=k.movie_id
inner join keywords kw
-- keywordmap with keyword 
on k.keyword_id=kw.keyword_id
where genre_name like 'Thriller' and 
kw.keyword_name like '%love%'
order by m.movie_id
limit 20;
 
```

### detecting duplicate with row_number window function 

```python
%%sql
with thiller_love as(
select 
        m.title, 
        m.movie_id,
        gen.genre_name ,
        kw.keyword_name ,
        row_number() over(partition by m.movie_id order by kw.keyword_name) as rank_movie

from movies as m
inner join genremap as g
on m.movie_id=g.movie_id
-- movie with genremap
inner join genres as gen 
on gen.genre_id= g.genre_id
-- genres with genremap
inner join keywordmap k
-- movie with keywordmap
on m.movie_id=k.movie_id
inner join keywords kw
-- keywordmap with keyword 
on k.keyword_id=kw.keyword_id
where genre_name like 'Thriller' and 
kw.keyword_name like '%love%'
order by m.movie_id
)
select *
from thiller_love 
where rank_movie>1
order by thiller_love.movie_id
```
